```python
#célula para a entrada de dados
d = [1,2,3,4,5] #número de dados que vamos usar (5)
va=[]
vt=[]
vv=[]
print()
print('Responda os valores com "." para separar as casa decimais, exemplo:(00.00)')
print()
for i in range(1,6):
      e={i}
     # print("Referente ao Mês ",[e])

      valiment = float(input(f"Quanto você gastou em Alimentação no mês{e}? "))
      va.append(valiment)

      valtransp = float(input(f"Quanto você gastou em Transporte no mês{e}? "))
      vt.append(valtransp)

      valvestu = float(input(f"Quanto você gastou em Vestuário no mês{e}? "))
      vv.append(valvestu)
      print()
```

```python
# célula para tratamento das séries temporais relacionadas a despesas;
import matplotlib.pyplot as plt
y = d
x1=va
x2=vt
x3=vv
```

```python
# Gerar os gráficos de linha das DESPESAS
plt.plot(y,x1,color='blue')
plt.plot(y,x2, color='brown')
plt.plot(y,x3, color='green')

plt.scatter(y,x1,color='red')
plt.scatter(y,x2,color='red')
plt.scatter(y,x3,color='red')

plt.title("Gráfico de despesas")
plt.ylabel('Despesas em R$')
plt.xlabel('Dias do mês')
plt.legend(['Alimentação','Transporte','Vestuário'])
```
